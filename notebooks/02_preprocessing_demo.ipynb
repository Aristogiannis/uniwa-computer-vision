# 02 — Preprocessing demo

Shows the percentile-normalisation and tiling steps on a single image. Useful for the report's preprocessing figure.

```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from cv_diffusion.preprocessing.normalize import percentile_normalize
from cv_diffusion.preprocessing.tile import TileSpec, iter_tiles
```

```python
# Choose any RGB image (xBD scenes are the typical input).
img_path = next(Path('../data/raw').rglob('*.png'), None)
assert img_path is not None, 'Drop at least one image into data/raw/ first.'
img = np.asarray(Image.open(img_path).convert('RGB'))
img.shape
```

```python
normed = percentile_normalize(img, lower=2.0, upper=98.0)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img); axes[0].set_title('raw'); axes[0].set_axis_off()
axes[1].imshow((normed*255).astype(np.uint8)); axes[1].set_title('percentile-normalized'); axes[1].set_axis_off()
plt.tight_layout(); plt.show()
```

```python
# Tile the normalised image and show the first nine tiles.
spec = TileSpec(size=512, stride=512, drop_partial=True)
tiles = [t for _, _, t in iter_tiles((normed*255).astype(np.uint8), spec)]
print(f'{len(tiles)} tiles produced from a {img.shape[1]}x{img.shape[0]} image')
fig, axes = plt.subplots(3, 3, figsize=(9, 9))
for ax, tile in zip(axes.flat, tiles[:9]):
    ax.imshow(tile); ax.set_axis_off()
plt.tight_layout(); plt.show()
```
